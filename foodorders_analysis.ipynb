```python
import pandas as pd
```

```python
df = pd.read_csv('foodorders.csv')
df.head()
```

```python
df.info()
```

```python
# Changing order date column to datetime
df['order_date'] = pd.to_datetime(df['order_date'])
```

## Top 10 City with highest sales

```python
sales_ranked = df.groupby('city')['price'].sum().reset_index() # Here price of orders placed is sales
sales_ranked.sort_values(by='price', ascending=False).reset_index(drop=True).head(10)
```

## Categories of items ordered by customers

```python
unique_items_categories = df['category'].unique()
print(unique_items_categories)
```

## Cities with most orders

```python
top_cities = df.groupby('city')['item_name'].count().reset_index()
top_cities.sort_values(by='item_name', ascending=False).reset_index(drop=True).head(10)
```

## Top customers in each city

```python
top_customers = df.groupby(['city', 'id_user'])['item_name'].count().reset_index()
top_customers['Rank'] = top_customers.groupby('city')['item_name'].rank(method='dense', ascending=False)
top_customers[top_customers['Rank']==1].reset_index(drop=True)
```

## Date-wise percentage change in sales

```python
new_df = df.groupby('order_date')['price'].sum().reset_index() # Here price is sales
new_df['previous_sales'] = new_df['price'].shift(1) # Here price is sales
new_df['Percentage_change'] = (new_df['price']-new_df['previous_sales'])/new_df['previous_sales']
new_df.round(2)
```
